```python
# 4
user_activity_events=[{"user_id":1, "event_type":"login", "duration_seconds":300},
                      {"user_id":2, "event_type":"logout", "duration_seconds":100},
                      {"user_id":3, "event_type":"purchase", "duration_seconds":-200},
                      {"user_id":4, "event_type":"register", "duration_seconds":300},
                      {"user_id":5, "event_type":"login", "duration_seconds":300},
                      {"user_id":6, "event_type":"purchase", "duration_seconds":200},
                      {"user_id":7, "event_type":"logout", "duration_seconds":300},
                      {"user_id":8, "event_type":"login", "duration_seconds":"non_numeric"},
                      {"user_id":9, "event_type":"register", "duration_seconds":200},
                      {"user_id":10, "event_type":"logout", "duration_seconds":300},
                      {"user_id":11, "event_type":"login", "duration_seconds":-300},
                      {"user_id":12, "event_type":"purchase", "duration_seconds":200},
                      {"user_id":13, "event_type":"logout", "duration_seconds":300},
                      {"user_id":14, "event_type":"login", "duration_seconds":300},
                      {"user_id":15, "event_type":"purchase", "duration_seconds":200}]

cleaned_events = []
for event in user_activity_events:
    duration = event.get("duration_seconds")
    if not isinstance(duration, (int, float)) or duration < 0:
        continue
    event["duration_minutes"] = duration / 60
    cleaned_events.append(event)
print(len(cleaned_events))
print(cleaned_events)
print(cleaned_events[:2])

# 5
event_counts = {}
for event in cleaned_events:
    event_type = event.get("event_type")
    if event_type:
        event_counts[event_type] = event_counts.get(event_type, 0) + 1
print(event_counts)

duration_means = {}
for event in cleaned_events:
    event_type = event.get("event_type")
    duration = event.get("duration_minutes")
    if event_type and isinstance(duration, (int, float)):
        if event_type not in duration_means:
            duration_means[event_type] = {"total_duration": 0, "count": 0}
        duration_means[event_type]["total_duration"] += duration
        duration_means[event_type]["count"] += 1
print({event_type: data["total_duration"] / data["count"] for event_type, data in duration_means.items()})

unique_users=set()
for event in cleaned_events:
    user_id = event.get("user_id")
    if user_id:
        unique_users.add(user_id)
print(len(unique_users))

rows = ["metric,key,value"]

for k,v in event_counts.items():
    rows.append(f"count,{k},{v}")

for k,d in duration_means.items():
    rows.append(f"average_duration_minutes,{k},{round(d['total_duration']/d['count'],4)}")

rows.append(f"unique_users,total,{len(unique_users)}")

csv_string = "\n".join(rows)
print(csv_string)
```
